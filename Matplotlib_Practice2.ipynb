```python
import matplotlib.pyplot as plt

#pie Chart 圓餅圖

#加入label
labels = 'A','B','C','D','E','F'

#設定每塊的大小
size = [33,52,12,17,62,48]

#圓餅圖使用plt.pie()
#size與labels都是上面設定的資料，autopct='%1.1f%%'是用來顯示百分比
plt.pie(size , labels = labels,autopct='%1.1f%%')

#讓圓餅圖比例相等加上
plt.axis('equal')
plt.show()
```

```python
#Separated 在pie裡面還有一個參數explode可以使用，我們直接來使用吧

import matplotlib.pyplot as plt

#pie Chart 圓餅圖

#加入label
labels = 'A','B','C','D','E','F'

#設定每塊的大小
size = [33,52,12,17,62,48]

#上面幾個.1,0,0,0,0,0 分別為 'A','B','C','D','E','F'，這是設定分開的數值(距離)
separated = (.4,0,.3,0,0,0)

#修改plt.pie()的內容新增一個explode參數
plt.pie(size , labels = labels,autopct='%1.1f%%',explode=separated) 

#等比例
plt.axis('equal')
plt.show()
```

```python
#With pandas 現在要加上我們之前講的pandas，來把pandas的資料做視覺化

import matplotlib.pyplot as plt
import pandas as pd

#先定義資料
data = {'names':['a','b','c','d','e'],
        'jan':[133,122,101,104,320],
        'feb':[122,132,144,98,62],
        'march':[64,99,32,12,65] }

#接著將資料設定為DataFrame資料格式，放到變數df
df = pd.DataFrame(data,columns=['names','jan','feb','march'])

#定義（總）大小
df['total'] = df['jan']+ df['feb']+df['march']

#定義顏色
colors = [(1,.4,.4),(1,.6,1),(.5,.3,1),(.7,.7,.2),(.6,.2,.6)]

#定義圓餅圖，上面定義的資料放進去
plt.pie( df['total'] ,
    labels = df['names'],
    colors = colors,
    autopct='%1.1f%%',
    ) 

#等比例
plt.axis('equal') 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

col_count = 3
bar_width = 0.2
index = np.arange(col_count)
A_scores = (553,536,548)
B_scores = (518,523,523)
C_scores = (613,570,588)
D_scores = (475,505,499)

#alpha是透明度
A = plt.bar(index,
           A_scores, 
           bar_width,
           alpha=.4,
           label="K") 
B = plt.bar(index+0.2,
            B_scores,
            bar_width,
            alpha=.4,
            label="C") 
C = plt.bar(index+0.4,
            C_scores,
            bar_width,
            alpha=.4,
            label="N") # x,y ,width
D = plt.bar(index+0.6,
            D_scores,
            bar_width,
            alpha=.4,
            label="F") # x,y ,width

#這裡我們用到了plt.text()定義文字
def createLabels(data):
    for item in data:
        height = item.get_height()
        plt.text(
            item.get_x()+item.get_width()/2., 
            height*1.05, 
            '%d' % int(height),
            ha = "center",
            va = "bottom",
        )
createLabels(A)
createLabels(B)
createLabels(C)
createLabels(D)

plt.ylabel("Mean score")
plt.xlabel("Subject")
plt.title("Test Scores by Contry")

#plt.legend()為右上角的圖。
#plt.xticks為底下的文字（為了至中所以+.3 / 2）
plt.xticks(index+.3 / 2 ,("Math","Reading","Science"))
plt.legend() 

plt.grid(True)
plt.show()
```
