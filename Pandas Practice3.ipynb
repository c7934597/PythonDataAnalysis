```python
'''
loc[]
再來要介紹loc[]的方法，用index的標籤來取出資料
'''

import pandas as pd
df = pd.read_csv('C:/Users/02445/AssetsImportCompleteSample.csv',index_col="ASSETNUM")
df.loc["Note032"]
```

```python
'''
loc[]不止可以一次選取一個(row)
將要的資料放到陣列內，就可以一次選取多個（row）資料了！
'''

df.loc[["Note032","Note033"]]
```

```python
#iloc[]是用index位置來取我們要的資料

df = pd.read_csv('C:/Users/02445/AssetsImportCompleteSample.csv')
df.iloc[4]
```

```python
df = pd.read_csv('C:/Users/02445/AssetsImportCompleteSample.csv',index_col="ASSETNUM")

#取位置0到3 index的資料
df.iloc[0:4]

#取位置0,4 index的資料
df.iloc[[0,4]]

'''
我們將index設定為Film，再來用iloc[]
可以看到其實它也是照index位置去取出資料的，並不是因為index是數字!
'''
```

```python
#想抓哪個index且哪個欄位的值

df.loc["Note031","ASSETTAG"]
```

```python
'''
rename()
再來我們來看一下要怎麼將index或是columns的名稱改變吧
'''

df = pd.read_csv('C:/Users/02445/AssetsImportCompleteSample.csv',index_col="ASSETNUM")
df.rename(columns={"ASSETTAG":"Release Date","BINNUM":"Revenue"},inplace=True)
df.head(3)
```

```python
'''
rename(index)
上面為改變columns，再來改變index
'''

df.rename(index={"Note031":"Doctor No",
                   "Note032":"Golden Eye",
                   "Note033":"Best bond Movie Ever"},inplace=True)
df.head(3)
```

```python
'''
drop()
刪除直接使用.drop()，裡面放入要刪除的labels，設定axis="columns"
'''

df.drop(labels=["Release Date","AUTOWOGEN"], axis="columns")
```

```python
#axis=1跟axis="columns"是一樣的！axis=0則是和asxis = "row"一樣。

df.drop(labels=["Note034","Note038"], axis=0)
```
