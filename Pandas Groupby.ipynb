```python
#一個簡單的例子，先定義col和data內容

col = ['class','name','hbd']
data = [['class0', 'user0', '1993-10-01'],
        ['class0', 'user1', '1992-10-02'],
        ['class1', 'user2', '1990-10-01'],
        ['class2', 'user3', '1983-10-03'],
        ['class1', 'user4', '1991-10-02'],
        ['class0', 'user5', '2001-10-03']]
frame = pd.DataFrame(data,columns=col)
frame
```

```python
#分組之後，可以用groups來看結果

frame_class = frame.groupby('class')
frame_class.groups
```

```python
#使用get_group()取出class1

frame_class.get_group("class1")
```

```python
import pandas as pd
import numpy as np

#定義一組資料集
df = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar',
                      'foo', 'bar', 'foo', 'foo'],
                   'B' : ['one', 'one', 'two', 'three',
                          'two', 'two', 'one', 'three'],
                   'C' : np.random.randn(8),
                   'D' : np.random.randn(8)})
df
```

```python
#計算總和

df.groupby('A').sum()
```

```python
#計算兩群組總和

df.groupby(['A','B']).sum()
```
