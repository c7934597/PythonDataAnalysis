```python
'''
標準化是使用了叫做preprocessing的函數，我們引入它。需要創建資料所以也引入numpy
'''
from sklearn import preprocessing
import numpy as np

'''
用train_test_split來把資料做切分（一邊為訓練一邊為測試資料）
用make_classification用來產生隨機的訓練資料
用SVC這個分類法當作例子
'''
from sklearn.model_selection import train_test_split
from sklearn.datasets.samples_generator import make_classification
from sklearn.svm import SVC

#需要繪圖
import matplotlib.pyplot as plt

#有兩部分的資料，因為我們設定了n_features=2
X,y = make_classification(n_samples=300,n_features=2,n_redundant=0,n_informative=2,
                          random_state=3,scale=100,n_clusters_per_class=1)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
#將資料切分，並且用SVC分類法進行訓練
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
clf = SVC(gamma='auto')
clf.fit(X_train,y_train)
```

```python
#最後用score()來查看訓練的分數如何
clf.score(X_test,y_test)
```

```python
#標準化：為了證明標準化能夠讓幫助資料訓練，我們使用preprocessing.scale()標準化X
X = preprocessing.scale(X)

#一樣區分資料，並且訓練
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
clf = SVC(gamma='auto')
clf.fit(X_train,y_train)

#看看得分多少
clf.score(X_test,y_test) #1.0

#X軸與Y軸的範圍縮小
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```
