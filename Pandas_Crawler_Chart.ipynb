```python
import pandas as pd
pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader as web
import datetime

#Remote Data Access 定義資料
start = datetime.datetime(2017, 1, 1)
end = datetime.datetime(2018, 1, 1)

#抓取yahoo上的股市資料
#上面這邊引入了yahoo上"F"也就是Finance的資料，並給予要抓取的開始時間start與結束時間end
data = web.DataReader("F", 'yahoo', start, end)

#資料太多縮成五筆
data.head()
```

```python
#with Numpy 用padas繪圖，若要用Numpy的資料呢

import numpy as np

#建立一筆隨機的資料np.random.randn(100, 4)，x為日期，columns設定為A,B,C,D
df = pd.DataFrame(np.random.randn(100, 4), index=pd.date_range('1/1/2017', periods=100), columns=list('ABCD'))

#將資料累加
df = df.cumsum()

df.plot()
```
