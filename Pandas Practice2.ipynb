```python
#資料篩選

df = pd.read_csv('C:/Users/02445/AssetsImportCompleteSample.csv')
df

#運算子運用
df["ASSETTAG"] == 66
```

```python
#過濾器

fliter = (df["ASSETTAG"] == 66)
df[fliter]
```

```python
#運算子運用
df["ASSETTAG"] != 66
```

```python
#運算子運用
df[df["ASSETTAG"] > 66]

#過濾日期
mask = df["CHANGEDATE"] <= "2012-09-22"
df[mask]
```

```python
#過濾不只一個條件

mask1 = df["ASSETNUM"] != "Note033"
mask2 = df["CHANGEDATE"] < "2012-09-22"

#條件和
df[(mask1 & mask2)]

#條件或
df[(mask1 | mask2)]
```

```python
mask = df["ASSETNUM"].isin(["Note037","Note040","Note042"])
df[mask]
```

```python
#isnull()如果值是空的就會顯示True

mask = df["CHANGEPMSTATUS"].isnull()
df[mask]
```

```python
#notnull()如果值不是空的就會顯示True

mask = df["CHANGEPMSTATUS"].notnull()
df[mask]
```

```python
#between()

df[df["CHANGEDATE"].between("2011-09-22","2012-09-22")]
```
