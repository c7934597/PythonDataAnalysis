```python
import pandas as pd

df = pd.read_csv('C:/Users/02445/AssetsImportCompleteSample.csv',parse_dates=["CHANGEDATE"])
df.head(10)
```

```python
#使用set_index設定index

df.set_index(keys = ["CHANGEDATE","ASSETNUM"],inplace=True)
df.head()
```

```python
#使用index.names查看有哪些index

df.index.names
```

```python
#使用type()查看資料的類型

type(df.index)
```

```python
#使用sort_index()排序資料，因為上面我們已經將index改變，所以排序也會依照CHANGEDATE以及ASSETNUM

df = pd.read_csv('C:/Users/02445/AssetsImportCompleteSample.csv',parse_dates=["CHANGEDATE"],index_col=["CHANGEDATE","ASSETNUM"])
df.sort_index(inplace=True)
df.head(10)
```

```python
#得到第0index的資料

df.index.get_level_values(0)
df.index.get_level_values("CHANGEDATE")
```

```python
#得到第1index的資料

df.index.get_level_values(1)
df.index.get_level_values("ASSETNUM")
```
